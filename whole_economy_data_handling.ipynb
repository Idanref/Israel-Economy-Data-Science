**Clearing data, removing duplicates and handling missing data.**

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)

#Load data
df = pd.read_csv('/content/drive/Othercomputers/ה-מחשב נייד שלי/Desktop/Python Project/Data/whole_economy_data.csv', index_col=0)

#Drop 2 rows with missing essence
df.dropna(inplace=True)

#Drop duplicates
df.drop_duplicates(inplace=True)

#New column of workplace seniority
df['workplace_seniority'] = df['year_of_resignation']-df['year_of_employment']

#Handling incorrectly entered data in columns year_of_employment and year_of_resignation (swap)
df.loc[df['workplace_seniority'] < 0,['year_of_employment', 'year_of_resignation']] = df.loc[df['workplace_seniority'] < 0, ['year_of_resignation', 'year_of_employment']].values

#Recalculate workplace seniority
df['workplace_seniority'] = df['year_of_resignation']-df['year_of_employment']

#Sort by year of workplace seniority
df = df.sort_values('workplace_seniority', ascending=True, ignore_index=True)

#Export data into csv.
df.to_csv('whole_economy_data.csv', encoding='utf-8-sig')
```
